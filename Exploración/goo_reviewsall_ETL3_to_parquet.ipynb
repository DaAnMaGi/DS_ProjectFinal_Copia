```python
import pandas as pd
import os
```

```python
# Lista para almacenar los dataframes de todos los estados
dataframes = []

# Funci√≥n para procesar cada estado y ruta
def procesar_archivos_json(state, ruta_directorio):
    for filename in os.listdir(ruta_directorio):
        if filename.endswith('.json'):
            try:
                df_temporal = pd.read_json(os.path.join(ruta_directorio, filename), lines=True)
                df_temporal['state'] = state
                dataframes.append(df_temporal)
            except ValueError as e:
                print(f"Error al procesar el archivo {filename}: {e}")
```

```python
state = 'AL'
ruta_directorio = '../Data/Google/reviews-estados/review-Alabama/review-Alabama'
dataframes = []

for filename in os.listdir(ruta_directorio):
    if filename.endswith('.json'):
        df_temporal = pd.read_json(os.path.join(ruta_directorio, filename), lines=True)
        dataframes.append(df_temporal)

df = pd.concat(dataframes, ignore_index=True)
df['state'] = state

file_name = f'../Data/Google/reviews-estados-parquet/revgoo_{state}.parquet'
df.to_parquet(file_name, index=False)
```
